```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd 
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://www.bankofengland.co.uk/news/prudential-regulation"

# Set up the Selenium driver
driver = webdriver.Chrome()

driver.get(url)
driver.implicitly_wait(5)
button = driver.find_element(By.XPATH,"//button[contains(@class,'cookie__button btn btn-default btn-neutral')]")
button.click()
driver.implicitly_wait(5)


lst = []
title_lst = []
d_lst = []

elements = driver.find_elements("xpath","//a[contains(@class, 'release release-regulation')]")
# title = driver.find_elements("xpath", "//h3[contains(@class, 'grid-view exclude-navigation')]")
date = driver.find_elements("xpath", "//time[contains(@class,'release-date')]")

    
# for t in title:
#     if t.text not in title_lst:
#         title_lst.append(t.text)
for ele in elements:
    if ele.get_attribute('href') not in lst:
        lst.append(ele.get_attribute('href'))
            
for d in date:
    if d.text not in d_lst:
        d_lst.append(d.text)
```

```python
# Unpack

final_d = list()
current_datetime = datetime.now()
current_month = current_datetime.strftime('%B')
current_year = str(current_datetime.year)

for i in d_lst:
    if current_year in i and current_month in i:
        final_d.append(i)
        
final_l = lst[:len(final_d)]
final_t = list()

for url in final_l:
    # Split the URL by "/"
    parts = url.split("/")
    # Get the last part of the URL
    text = parts[-1]
    # Print the extracted text
    text = text.replace("-", " ").title()
    final_t.append(text)

df = pd.DataFrame({"Organisation" : ["PRA" for i in final_d],
                 "Title" : final_t,
                 "Link" : final_l,
                 "Date" : final_d})


current = os.getcwd().replace("\\", "/")
file_path = "/Downloads/"
# Save to Downloads Folder
save_string = "RPA_updates.xlsx"
save_string = current + file_path + save_string # Comment out to save to directory file is in
df.to_excel(save_string, index = False)
```
